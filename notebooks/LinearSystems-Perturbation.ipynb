```python
import matplotlib.pyplot as plt

sizes = [
    100,
    250,
    500,
    750,
    1000
]

dss_size = [
    1.4e-5,
    2.3e-5,
    5.1e-5,
    2.9e-4,
    3.7e-3
]

proxy_size = [
    1.0e-4,
    5.5e-5,
    3.3e-5,
    7.8e-5,
    1.5e-3
]

plt.figure(figsize=[2, 2], dpi=300)
plt.scatter(sizes, dss_size, color='blue', marker='+')
plt.scatter(sizes, proxy_size, color='red', marker='+')
plt.yscale('log')
plt.xlabel(r'$n$')
plt.ylabel(r'$1 - Corr$')
plt.legend(['DSS', 'Proxy'])

plt.ylim([10**(-6), 10**(-1)])

plt.show()
```

```python
noises = [
    0.001,
    0.01,
    0.1,
    1,
    10
]

dss_noise = [
    1.1e-4,
    7.8e-5,
    1.9e-4,
    5.1e-2,
    0.9996
]

proxy_noise = [
    3.8e-5,
    3.7e-5,
    8.1e-5,
    0.11,
    0.995
]

plt.figure(figsize=[2, 2], dpi=300)
plt.scatter(noises, dss_noise, color='blue', marker='+')
plt.scatter(noises, proxy_noise, color='red', marker='+')
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$\tau$')
plt.ylabel(r'$1 - Corr$')
plt.legend(['DSS', 'Proxy'])

plt.xlim([10**(-3.5), 10**(1.5)])
plt.ylim([10**(-5), 10**(0.5)])

plt.show()
```
